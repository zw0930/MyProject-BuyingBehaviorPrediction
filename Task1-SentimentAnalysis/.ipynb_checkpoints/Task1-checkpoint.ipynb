## Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
# Import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import datetime
```

```python
# Scrapping the reviews data from the website

base_url = "https://www.airlinequality.com/airline-reviews/british-airways"

pages = 37      # Page number, Total = 37
page_size = 100 

# Loop over pages

# Find all the categories in the rating table

# Review rating headers
tab_kwd = []

for i in range(1, pages + 1):

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    
    # Find all the review rating headers
    for para in parsed_content.find_all("table", {"class":"review-ratings"})[1:]: 
        for para1 in para.find_all("td", {"class":re.compile("review-rating-header*")}):
            if para1.get_text() not in tab_kwd:
                tab_kwd.append(para1.get_text())   

#print(tab_kwd)
print(f"Found all the categories!")
    
```

```python
# Other elements for each review

# Header
title = []
# Subheader
author = []
location = [] # Need tp clean up 
date_published = []
rating = []
# Text
review = []
# Table 
# Dictionary containing all the data in the tables 
table = {tab_kwd[i]: [] for i in range(len(tab_kwd))} 

# Collect data
for i in range(1, pages + 1):
    
    from IPython.display import clear_output
    clear_output(wait=False)
    print(f"Scraping page {i}/{pages}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    
    # Subheader (name, location, date published)
    for para in parsed_content.find_all("div", {"class":"body"}):
        title.append(para.find("h2", {"class":"text_header"}).get_text())
        para1 = para.find("h3", {"class":"text_sub_header userStatusWrapper"})
        author.append(para1.find("span", {"itemprop":"name"}).get_text())
        location.append(para1.get_text())
        date_published.append(para1.find("time", {"itemprop":"datePublished"}).get_text())
    
    # Rating (Best rating is 10)
    
    for para in parsed_content.find_all("div", {"class":"rating-10"})[1:]:
        if para.find("span", {"itemprop":"ratingValue"}):
            rating.append(para.find("span", {"itemprop":"ratingValue"}).get_text())
        else:
            rating.append('None')
        
    # Review text
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        review.append(para.get_text())
    
    # Review rating table   
    for para in parsed_content.find_all("table", {"class":"review-ratings"})[1:]:
        for key in tab_kwd:
            if para.find("td", text=key):
                if para.find("td", text=key).find_parent("tr").find("td",{"class":"review-value"}):
                    table[key].append(para.find("td", text=key).find_parent("tr").find("td",{"class":"review-value"}).get_text())
                elif para.find("td", text=key).find_parent("tr").find("td",{"class":"review-rating-stars stars"}):
                    para1 = para.find("td", text=key).find_parent("tr").find("td",{"class":"review-rating-stars stars"})
                    if para1.find("span",{"class":"star fill"}):
                        table[key].append(para1.find_all("span",{"class":"star fill"})[-1].get_text())
                    else:
                        table[key].append('0')
            else:
                table[key].append("None")    
    
    #print(f"   ---> {len(review)} total reviews")
    #print(f"   ---> {len(table['Aircraft'])} total aircraft")
    
print(f"Scraping finished successfully!")
    
```

```python
# Save the reviews in DataFrame

df1 = pd.DataFrame(
    {'Title': title,
     'Author': author,
     'Country': location,
     'Date Published': date_published,
     'Rating': rating,
     'Review': review
    })
#for key in tab_kwd:
#    print(key, len(table[key]))
df2 = pd.DataFrame.from_dict(table, orient = 'columns')
df = pd.concat([df1, df2], axis=1)
df.head()
```

## Cleaning data

Review the data in each column and write the cleaned data into a csv file for the analysis.

Clean the data in each column

```python
# Title: Remove the double quoataion marks in the titles and set all words to lower case

df['Title'] = df['Title'].str.replace('"','')
df['Title'] = df['Title'].str.lower()
df.head()

# Author: Format names as 'initial of first name + last name', e.g. A Xyz

s = list(df['Author'].str.split())
df['Author'] = list(map(lambda x: x[0][:1] + ' ' + x[-1], s))

# Country: Extract only the country names from the text

df['Country'] = df['Country'].str.replace(r'[^(]*\(|\)[^)]*', '')

# Date Published: Covert to datetime

s = df['Date Published'].str.split()
for i in range(len(s)):
    s[i][0] = re.findall(r'\d+', s[i][0])[0]
    s[i] = str(s[i][0]) + ' ' + str(s[i][1]) + ' ' + str(s[i][2])

df['Date Published'] = s
#df['Date Published'] = list(map(lambda x: datetime.datetime.strptime(x, '%d %B %Y'), df['Date Published']))
df['Date Published'] = [datetime.datetime.strptime(x, '%d %B %Y') if x != 'None' else 'None' for x in list(df['Date Published'])]
```

```python
# Rating: Convert digital string to int if not None

df['Rating'] = [int(i) if i != 'None' else 'None' for i in list(df['Rating'])]

# Review: Keep only the text after '|' and remove the whitespace from the beginning and the end of the string

df['Review'] = [s[s.find('|') + 1:].strip() if s.find('|') > -1 else s[s.find('|') + 1:].strip() for s in list(df['Review'])]

# Aircraft: Looking good for now

# Type Of Traveller: Looking good for now

# Seat Type: Looking good for now

# Route: Looking good for now, will split into Departure and Arrive in the analysis later

# Date Flown: Convert to datetime 
df['Date Flown'] = [datetime.datetime.strptime(x, '%B %Y') if x != 'None' else 'None' for x in list(df['Date Flown'])]
      
```

```python
# Columns from the tables: Convert to int if not None

df['Seat Comfort'] = [int(i) if i != 'None' else 'None' for i in list(df['Seat Comfort'])]
df['Cabin Staff Service'] = [int(i) if i != 'None' else 'None' for i in list(df['Cabin Staff Service'])]
df['Food & Beverages'] = [int(i) if i != 'None' else 'None' for i in list(df['Food & Beverages'])]
df['Inflight Entertainment'] = [int(i) if i != 'None' else 'None' for i in list(df['Inflight Entertainment'])]
df['Ground Service'] = [int(i) if i != 'None' else 'None' for i in list(df['Ground Service'])]
df['Wifi & Connectivity'] = [int(i) if i != 'None' else 'None' for i in list(df['Wifi & Connectivity'])]
df['Value For Money'] = [int(i) if i != 'None' else 'None' for i in list(df['Value For Money'])]

# Recommended: Replace 'yes' with 1, 'no' with 0
df['Recommended'] = df['Recommended'].map({'yes': 1, 'no': 0})
df.head()
```

```python
# Write the cleaned data to csv

df.to_csv("/Users/a2046/Documents/_Job/BritishAirways_Forage/data/BA_reviews_cleaned.csv")
```

#### Ratings of different aspects of the service
- Utilizing the star reviews in the tables, we can have an overview on which aspects are the most satified and which ones are the least satisfied for further investigation.
- The rating is from 0 to 5 (stars) and still categorized into 3 levels for the straightforwant high-level evaluation. The levels are: 0 to 2, negative; 3, neutral; 4 to 5, positive.

```python
def rating_percent_star(df):
    col = df.columns[0]
    pos, neu, neg, none = 0, 0, 0, 0
    for i in df[col]:
        if isinstance(i, int):
            if (i >= 0) & (i <= 2):
                pos += 1
            elif i == 3:
                neu += 1
            elif (i >= 4) & (i <= 5):
                neg += 1
        else: 
            none +=1
    l = list([pos, neu, neg, none])
    l1 = [round(x / len(df) * 100) for x in l]
    
    if  sum(l1) == 100:
        return l1
    else:
        # If the rounded int for percentages doesn't add up to 100, change the number with the smallest error
        tmp = [x / len(df) * 100 for x in l]
        tmp = list(abs(np.array(tmp) - np.array(l1)) / np.array(tmp))
        m = 0
        for i in tmp:
            if i == min(tmp):
                break
            else:
                m += 1
        l1[m] = l1[m] - (sum(l1) - 100)
        return l1
```

```python
# 0: Seat Comfort
# 1: Cabin Staff Service
# 2: Food & Beverages
# 3: Inflight Entertainment
# 4: Ground Service
# 5: Wifi & Connectivity
# 6: Value For Money

plot_rev0 = rating_percent_star(df[['Seat Comfort']]) 
print(plot_rev0, sum(plot_rev0))

plot_rev1 = rating_percent_star(df[['Cabin Staff Service']]) 
print(plot_rev1, sum(plot_rev1))

plot_rev2 = rating_percent_star(df[['Food & Beverages']]) 
print(plot_rev2, sum(plot_rev2))

plot_rev3 = rating_percent_star(df[['Inflight Entertainment']]) 
print(plot_rev3, sum(plot_rev3))

plot_rev4 = rating_percent_star(df[['Ground Service']]) 
print(plot_rev4, sum(plot_rev4))

plot_rev5 = rating_percent_star(df[['Wifi & Connectivity']]) 
print(plot_rev5, sum(plot_rev5))

plot_rev6 = rating_percent_star(df[['Value For Money']]) 
print(plot_rev6, sum(plot_rev6))
```
